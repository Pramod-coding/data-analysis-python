# Set Operations


```python
# Creating a set
fruits = {"apple", "banana", "cherry"}
numbers = set([1, 2, 3, 4, 5])
fruits
numbers
```

```python
#Uninon sets 
#To get the union of two sets, you can use the union() method or the | operator.
st={'A','B','C'}
st1={'D','E','F'}
st2={'G','H','I'}
st|st1|st2
```

```python
#Intersection of Sets:
# To find the intersection of two sets, you can use the intersection() method or the & operator.st={'A','B','C'}
st3 = {1, 2, 3}
st4 = {3, 4, 5}
```

```python
st3 & st4
```

```python
#Difference between Sets
# To get the difference between two sets (elements present in one set but not the other),
#you can use the difference() method or the - operator.

st3 - st4
```

```python
st1.difference(st2)
```

```python
#Symmetric Difference between Sets:
# The symmetric difference of two sets contains elements that are present in either set but not in 
#both. You can use the symmetric_difference()method or the ^ operator.

st3 ^ st4
```

```python
st.symmetric_difference(st2)
```

```python
# Subset:
#To check if one set is a subset of another set (i.e., all elements of the first set are also
#present in the second set), you can use the issubset() method or the <= operator.
st5={7,8,9,10,11,12,13,14,15}
st6={9,10,11,12,13,14}

st6<=st5
```

```python
st5.issubset(st6)
```

```python
#Superset:
#To check if one set is a superset of another set (i.e., the second set contains all elements
#of the first set), you can use the issuperset() method or the >= operator.
st6>=st5

```

```python
st5.issuperset(st6)
```

```python
#Disjoint:
#To check if two sets are disjoint (i.e., they have no common elements), you can use the
#isdisjoint() method
st5.isdisjoint(st6)
```
