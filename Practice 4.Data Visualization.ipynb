```python
#Using Hexadecimal RGB Values:
#You can specify colors using hexadecimal RGB values. The format is '#RRGGBB', where RR, GG, and
#BB are two-digit hexadecimal values representing the intensity of red, green, and blue, respectively.
plt.plot(Salaries[0], c='#0000FF') #In this example, #0000FF corresponds to the color blue. 
#The first two digits 00 represent the intensity of red (0), the next two digits 00 represent
#the intensity of green (0), and the last two digits FF represent the intensity of blue (255)
```

```python
#Using RGB Tuples:
#You can also specify colors using RGB tuples, where each element of the tuple represents
#the intensity of red, green, and blue, respectively. The values range from 0 to 1.
plt.plot(Salaries[0], c=(0.8, 0.4, 0.1))
```

```python
#Using Short Color Codes:
#Matplotlib also provides short color codes for some common colors, such as 'b' for blue, 
#'g' for green, 'r' for red, 'c' for cyan, 'm' for magenta, 'y' for yellow, 'k' for black,
#and 'w' for white.
plt.plot(Salaries[0], c='r')
```

```python
#Dictionaries:
#In Matplotlib, dictionaries are used to store various configuration settings for different plot
#elements. These dictionaries are often referred to as 
#"rcParams" (short for "Runtime Configuration Parameters"). 
#These parameters define the default styles and behavior of Matplotlib plots and can
#be customized to modify the appearance of the plots to suit your preferences.
```

```python
#figure.figsize: Specifies the default width and height of the figure in inches.
%matplotlib inline
plt.rcParams['figure.figsize'] = 8, 6
plt.plot(Salaries[0])
plt.title('figure.figsize')
```

```python
#figure.dpi: Sets the dots-per-inch (resolution) for saving figures.
plt.rcParams['figure.dpi'] = 100
plt.plot(Salaries[0])
plt.title('figure.dpi')
```

```python
#font.family: Specifies the default font family for text elements in the plot.
plt.rcParams['font.family'] = 'serif'
plt.plot(Salaries[0])
plt.title('font.family')
```

```python
#font.size: Sets the default font size for text elements.
plt.rcParams['font.size'] = 12
plt.plot(Salaries[0])
plt.title('font.size')
```

```python
#axes.grid: Specifies whether to display grid lines in the plot.
plt.rcParams['axes.grid'] = True
plt.plot(Salaries[0])
plt.title('axes.grid')
```

```python
#lines.linewidth: Sets the default line width for plot lines.
plt.rcParams['lines.linewidth'] = 2.5
plt.rcParams['axes.grid'] = False
plt.plot(Salaries[0])
plt.title('lines.linewidth')
```

```python
#lines.marker: Specifies the default marker style for data points in line plots.
plt.rcParams['lines.marker'] = 'o'
plt.plot(Salaries[0])
plt.title('lines.marker')
```

```python
#Location Of Legend:
#'best': Automatically choose the best location to avoid overlapping with the plot elements.
plt.legend(loc = 'best')
#'upper right': Place the legend in the upper-right corner of the plot.
plt.legend(loc = 'upper right')
#'upper left': Place the legend in the upper-left corner of the plot.
plt.legend(loc = 'upper left')
#'lower right': Place the legend in the lower-right corner of the plot.
plt.legend(loc = 'lower right')
#'lower left': Place the legend in the lower-left corner of the plot.
plt.legend(loc = 'lower left')
#'right': Place the legend in the right-center of the plot.
plt.legend(loc = 'right')
#'center left': Place the legend in the center-left of the plot.
plt.legend(loc = 'center left')
#'center right': Place the legend in the center-right of the plot.
plt.legend(loc = 'center right')
#'center': Place the legend in the center of the plot.
plt.legend(loc = 'center')
```

```python
# bbox_to_anchor : is used in conjunction with the loc parameter in the legend() function of
#Matplotlib to further customize the position of the legend relative to a specific point on the plot.
plt.plot(Salaries[0], c='Green', ls = '--', marker = 's', ms = 7, label = Players[0])
plt.plot(Salaries[1], c='Blue', ls = '-', marker = 'o', ms = 5, label = Players[1])
plt.plot(Salaries[2], c='purple', ls = '-.', marker = '^', ms = 8, label = Players[2])
plt.plot(Salaries[3], c='Red', ls = ':', marker = 'd', ms = 8, label = Players[3])
plt.xticks(list(range(0,10)), Seasons,rotation='vertical')
plt.legend(loc = 'upper right',bbox_to_anchor=(1.15,1))
```
